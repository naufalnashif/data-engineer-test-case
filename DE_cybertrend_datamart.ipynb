#Import Dependencies

```python
import pandas as pd
from datetime import datetime
```

# Load Data SD Coklat, SD Jingga, SD Ungu

```python
data_sdCoklat = pd.read_csv("/content/SD_Coklat-transform_Ready.csv")
data_sdJingga = pd.read_csv("/content/SD_jingga-transform_Ready.csv")
data_sdUngu = pd.read_csv("/content/SD_ungu-transform_Ready.csv")
```

## Concat dataframe

```python
# Gabungkan DataFrame menggunakan pd.concat
dataMart = pd.concat([data_sdCoklat, data_sdJingga, data_sdUngu], ignore_index=True)
dataMart
```

## Buat Email

```python
# Fungsi untuk membuat email
def create_email(row):
    first_name_initial = row['fullName'].split()[0][0].lower()
    last_name = row['fullName'].split()[-1].lower()
    email = f"{first_name_initial}.{last_name}@sekolah.edu.id"
    return email

dataMart['email'] = dataMart.apply(create_email, axis=1)


# duplicate_emails = dataMart[dataMart.duplicated(subset='email', keep=False)]['email']
# print("Emails with Duplicates:")
# print(duplicate_emails)

# # Menambahkan sufix unik jika ada duplikasi
# dataMart['email'] = dataMart.groupby('email').cumcount().astype(str).replace('0', '') + '_' + dataMart['email'].astype(str)
```

## Buat kolom ID

```python
dataMart['ID'] = range(1, len(dataMart) + 1)
```

## Standarisasi NIS

```python
dataMart['Std-NIS'] = dataMart['NIS'] + '-' + dataMart['school'].str.split().str[1]
dataMart
```

## Reorder columns

```python
dataMart = dataMart[['ID','Std-NIS', 'NIS', 'school', 'fullName','email', 'gender', 'grade', 'POB', 'birth', 'age', 'caturWulan', 'subjects', 'midTest', 'finalTest', 'mean', 'finalScore']]
dataMart
```

```python
dataMart.to_csv("dataMart.csv", index=False)
dataMart.to_excel("dataMart.xlsx", index=False)
```
