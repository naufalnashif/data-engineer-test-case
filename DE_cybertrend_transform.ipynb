# Import Dependencies

```python
import pandas as pd
from datetime import datetime
```

#Data SD Coklat

## Load data

```python
sd_coklat = pd.read_excel('SD Coklat.xlsx', sheet_name=None)
```

## Transform Data SD Coklat

### Copy data 1

```python
data_coklat = sd_coklat.copy()
```

```python
sheets_name = list(data_coklat.keys())
```

```python
sheets_name.remove('Student')
```

```python
sheets_name
```

### Gabungkan semua sheets

```python
# Menggabungkan nilai pelajaran berdasarkan 'StudentID'
data_coklat_combined = None

for subject_sheet in sheets_name:
    # Menambahkan kolom 'Pelajaran' pada setiap dataframe
    data_coklat[subject_sheet]['Pelajaran'] = subject_sheet

    # Menggabungkan dataframe berdasarkan 'StudentID'
    if data_coklat_combined is None:
        data_coklat_combined = data_coklat[subject_sheet]
    else:
        data_coklat_combined = pd.concat([data_coklat_combined, data_coklat[subject_sheet]], ignore_index=True)

# Menggabungkan dengan dataframe 'Student' berdasarkan 'StudentID'
data_coklat_combined = pd.merge(data_coklat['Student'], data_coklat_combined, on='StudentID')

# Mengisi kolom 'Rata-Rata' dengan nilai rata-rata (Mid Test + Final Test)/2
data_coklat_combined['Rata-Rata'] = (data_coklat_combined['Mid Test'] + data_coklat_combined['Final Test']) / 2
```

### Copy data 2

```python
data_coklat_combined_2 = data_coklat_combined.copy()
```

### Tambahkan kolom full_name

```python
#Tambahkan kolom full name
data_coklat_combined_2['Full Name'] = data_coklat_combined_2['First Name'] + ' ' + data_coklat_combined_2['Middle Name'].fillna('') + ' ' + data_coklat_combined_2['Lastname'].fillna('')
```

### Tambahkan tanggal lahir, jadikan datetime, dan tambahkan kolom Umur

```python

data_coklat_combined_2['Tanggal Lahir'] = pd.to_datetime(data_coklat_combined_2['YOB'].astype(str) + '-' + data_coklat_combined_2['MOB'].astype(str) + '-' + data_coklat_combined_2['DOB'].astype(str), errors='coerce')

# Hitung umur per hari ini
data_coklat_combined_2['Umur'] = (datetime.now() - data_coklat_combined_2['Tanggal Lahir']).astype('<m8[Y]').astype(int)
```

### Tambahkan kolom Nilai Mutu A,B atau C

```python

# Tambahkan kolom 'Nilai Mutu' berdasarkan ketentuan
data_coklat_combined_2['Nilai Mutu'] = pd.cut(
    data_coklat_combined['Rata-Rata'],
    bins=[0, 49, 69, 84, 100],
    labels=['D', 'C', 'B', 'A'],
    right=True
)
```

```python
data_coklat_combined_2[data_coklat_combined_2['Rata-Rata'] == 50]
```

### Tambahkan kolom School

```python
data_coklat_combined_2['School'] = 'SD Coklat'
data_coklat_combined_2
```

### Reorder Columns

```python
# Reorder columns
column_order = ['StudentID', 'School', 'Full Name', 'First Name', 'Middle Name', 'Lastname', 'Gender', 'Grade', 'POB', 'DOB', 'MOB', 'YOB', 'Tanggal Lahir', 'Umur','Absence (days)', 'Term', 'Pelajaran','Mid Test', 'Final Test', 'Rata-Rata', 'Nilai Mutu']

data_coklat_combined_2 = data_coklat_combined_2[column_order]
```

### Ganti nama kolom

```python
# Ganti nama kolom
data_coklat_combined_2 = data_coklat_combined_2.rename(columns={
    'StudentID': 'NIS',
    'Full Name': 'fullName',
    'First Name': 'firstName',
    'Middle Name': 'middleName',
    'Lastname': 'lastName',
    'Gender': 'gender',
    'Grade': 'grade',
    'POB': 'POB',
    'DOB': 'DOB',
    'MOB': 'MOB',
    'YOB': 'YOB',
    'Tanggal Lahir': 'birth',
    'Umur': 'age',
    'Absence (days)': 'absence',
    'Term': 'caturWulan',
    'Pelajaran': 'subjects',
    'Mid Test': 'midTest',
    'Final Test': 'finalTest',
    'Rata-Rata': 'mean',
    'Nilai Mutu' : 'finalScore',
    'School':'school'
})
```

```python
data_coklat_combined_2
```

## Final dataset SD Coklat

```python
data_coklat_ready = data_coklat_combined_2.loc[:, ['NIS', 'school', 'fullName', 'gender', 'grade', 'POB', 'birth', 'age', 'caturWulan', 'subjects', 'midTest', 'finalTest', 'mean', 'finalScore']]
```

```python
data_coklat_ready
```

```python
data_coklat_ready.to_csv('SD_Coklat-transform_Ready.csv', index=False)
data_coklat_ready.to_excel('SD_Coklat-transform_Ready.xlsx', index=False)
```
